```python
from numpy import random
import networkx as nx
def average_distance_from_sample(G, n):
    """
    Return the average distance between ``n`` randomly selected
    pairs of points from the graph ``n``.

    INPUT:

    ``G`` - Sage graph.
    ``n`` - positibe integer (sample size)

    OUTPUT:

    floating point real number.
    """
    
    ls = G.vertices()
    x1 = sample(ls, n)
    x2 = sample(ls, n)
    dist = []
    for i in range(n):
        a, b= x1[i], x2[i]
        dist.append(G.shortest_path_length(a,b))
    avg = float(sum(dist)/n)
    return avg
```
